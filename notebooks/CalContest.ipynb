```python
# 这是关于DPS排名的一个比赛的分析代码

stdList = ["无方", "田螺", "大师", "焚影", "藏剑", "花间", "衍天", "凌雪", "霸刀", "苍云", "气纯", "剑纯", "冰心", "丐帮", "鲸鱼", "傲血", "毒经", "蓬莱", "莫问"]
std = {}
for i in range(len(stdList)):
    std[stdList[i]] = i+1

```

```python
ansList = """无方→田螺→蓬莱→霸刀→衍天→焚影→鲸鱼→花间→冰心→苍云→傲血→凌雪→丐帮→气纯→藏剑→剑纯→大师→莫问→毒经	花姐
无方→田螺→衍天→莫问→气纯→焚影→蓬莱→鲸鱼→毒经→藏剑→剑纯→冰心→花间→霸刀→苍云→傲血→丐帮→凌雪→大师	天乙玄晶
无方→田螺→衍天→蓬莱→焚影→鲸鱼→丐帮→苍云→霸刀→冰心→毒经→藏剑→剑纯→莫问→花间→大师→气纯→傲血→凌雪	220
无方→衍天→田螺→霸刀→丐帮→焚影→冰心→大师→蓬莱→毒经→花间→鲸鱼→莫问→凌雪→苍云→傲血→藏剑→剑纯→气纯	招财
无方→衍天→田螺→莫问→蓬莱→鲸鱼→剑纯→傲血→凌雪→霸刀→焚影→丐帮→苍云→气纯→冰心→藏剑→毒经→大师→花间	白菜头
无方→丐帮→衍天→田螺→焚影→苍云→霸刀→冰心→鲸鱼→蓬莱→藏剑→剑纯→凌雪→毒经→莫问→傲血→花间→气纯→大师	小青
无方→田螺→苍云→焚影→霸刀→丐帮→气纯→鲸鱼→藏剑→莫问→蓬莱→凌雪→衍天→傲血→花间→剑纯→毒经→大师→冰心	YNLZ
无方→田螺→藏剑→丐帮→蓬莱→霸刀→衍天→焚影→气纯→剑纯→冰心→毒经→花间→鲸鱼→莫问→苍云→傲血→大师→凌雪	陆寻寻
无方→田螺→毒经→冰心→莫问→剑纯→傲血→藏剑→霸刀→蓬莱→凌雪→衍天→苍云→丐帮→焚影→鲸鱼→气纯→花间→大师	唐简@破阵子"""

ans = {}
ansRaw = {}
a1 = ansList.split('\n')
for line in a1:
    name = line.split('\t')[1]
    ansRaw[name] = line.split('\t')[0]
    res = line.split('\t')[0].split("→")
    ansSingle = []
    for item in res:
        ansSingle.append(std[item])
    ans[name] = ansSingle
```

```python
print(ans)
```

```python
def calculRev(s):
    L = len(s)
    rev = 0
    for i in range(L-1):
        for j in range(i+1, L):
            if s[i] > s[j]:
                rev += 1
    return rev

ansRev = {}
for line in ans:
    ansRev[line] = calculRev(ans[line])

for line in ans:
    print(line, '\t', ansRev[line], '\t', ansRaw[line])
```
