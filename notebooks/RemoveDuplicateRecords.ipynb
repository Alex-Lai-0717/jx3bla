```python
# 移除重复的记录。在6.1.0版本附近生效。

import numpy as np
import pymysql
import configparser
import json
import time
import hashlib
```

```python
ip = "139.199.102.41"
config = configparser.RawConfigParser()
config.readfp(open('../settings.cfg'))

dbname = config.get('jx3bla', 'username')
dbpwd = config.get('jx3bla', 'password')
db = pymysql.connect(ip,dbname,dbpwd,"jx3bla",port=3306,charset='utf8')
cursor = db.cursor()
```

```python
edition = "5.1.0"
mapdetail = "25人普通达摩洞"

sql = """SELECT * FROM ActorStat WHERE edition>='%s' AND win=1"""%edition
cursor.execute(sql)
result = cursor.fetchall()
```

```python
def hashGroupNameOnly(l, date, boss, map):
    nameList = []
    for line in l:
        nameList.append(line)
    nameList.sort()
    hashStr = date + boss + "".join(nameList)
    hashres = hashlib.md5(hashStr.encode(encoding="utf-8")).hexdigest()
    return hashres

def getHash(rec):
    stat = rec[7].replace("'",'"')
    Jstat = json.loads(stat)

    namelist = []
    for line in Jstat["effectiveDPSList"]:
        namelist.append(line[0])

    hash1 = hashGroupNameOnly(namelist, rec[2], rec[1], rec[3])
    
    if "雷火丰" in namelist and rec[2] == "2021-01-29":
        print(rec)
        print(hash1)
    
    return hash1
```

```python
d = {}
sumRemove = 0
for i in range(len(result)):
    line = result[i]
    hash1 = getHash(line)
    if hash1 not in d:
        d[hash1] = i
    else:
        j = d[hash1]
        print(j, i)
        if result[j][4] < result[i][4]:
            print("Remove duplicate: %s, hash = %s"%(j, result[j][5]))
            remove = result[j][5]
            d[hash1] = i
        else:
            print("Remove duplicate: %s, hash = %s"%(i, result[i][5]))
            remove = result[i][5]
        sumRemove += 1
        sql = """DELETE FROM ActorStat WHERE hash = "%s" """%remove
        cursor.execute(sql)
        
print(sumRemove)
print(len(result))
```
