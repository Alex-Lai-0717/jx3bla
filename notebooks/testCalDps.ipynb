```python
def getColor(occ, light = 0):
    if occ[-1] in ['d', 't', 'h', 'p', 'm']:
        occ = occ[:-1]
    colorDict = {"0": (0, 0, 0), 
                 "1": (210, 180, 0),#少林
                 "2": (127, 31, 223),#万花
                 "4": (56, 175, 255),#纯阳
                 "5": (255, 127, 255),#七秀
                 "3": (160, 0, 0),#天策
                 "8": (255, 255, 0),#藏剑
                 "9": (205, 133, 63),#丐帮
                 "10": (253, 84, 0),#明教
                 "6": (63, 31, 159),#五毒
                 "7": (0, 133, 144),#唐门
                 "21": (180, 60, 0),#苍云
                 "22": (100, 250, 180),#长歌
                 "23": (71, 73, 166),#霸刀
                 "24": (195, 171, 227),#蓬莱
                 "25": (161, 9, 34),#凌雪
                 "211": (166, 83, 251),#衍天
                }
    if occ not in colorDict:
        occ = "0"
    a = colorDict[occ][0]/255
    b = colorDict[occ][1]/255
    c = colorDict[occ][2]/255
    if light == 1:
        a = (a + 1) / 2
        b = (b + 1) / 2
        c = (c + 1) / 2
    elif light == 2:
        n = a + b + c
        if n < 1.5:
            return (1, 1, 1)
        else:
            return (0, 0, 0)
    return (a, b, c)
```
