```python
f = open("raw.txt", "r", encoding='utf-8')
sumd = 0
n = 0
res = []
for line in f:
    d = float(line.split(':')[0])
    name = line.split(':')[1].strip('"').strip('\n')
    sumd += d
    n += 1
    res.append([name, d])
avgd = sumd / n
for line in res:
    std = line[1] / avgd * 100
    ale = line[1] / avgd * 90
    thr = line[1] / avgd * 80
    print(line[0], std, ale, thr)


```
